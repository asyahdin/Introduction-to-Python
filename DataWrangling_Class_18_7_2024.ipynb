```python
#first columns to find missing value
#find min find max

mean_norm = mydata['normalized-losses'].mean() #calculate mean value
mean_norm
```

```python
mean_norm = mydata['normalized-losses'].astype('float').mean() #convert to float
mean_norm #got the value of normalized-losses
```

```python
mydata['normalized-losses'].replace(np.nan, mean_norm, inplace=True) #replace missing value with mean value

mydata.isnull().sum() #check missing value
```

```python
mydata.describe(include='all') #describe all columns
```

```python

mydata['num-of-doors'].value_counts() #count how many value
```

```python
#most freq values is num-of-door = four
#this is how to count max num

mydata['num-of-doors'].value_counts().idxmax() #count how many value
```

```python
#replace missing value of the num-of-door

mydata['num-of-doors'].replace(np.nan, mydata['num-of-doors'].value_counts().idxmax(), inplace=True) #replace missing value with mean value

mydata.isnull().sum() #check missing value
```

```python
mean_bore = mydata['bore'].astype('float').mean() #convert to float
mean_bore #got the value of normalized-losses
```

```python
mydata['bore'].replace(np.nan, mean_bore, inplace=True) #replace missing value with mean value

mydata.isnull().sum() #check missing value
```

```python
mean_stroke = mydata['stroke'].astype('float').mean() #for stroke

mean_horse = mydata['horsepower'].astype('float').mean() #for horse

mean_peak = mydata['peak-rpm'].astype('float').mean() #for peak

mydata['stroke'].replace(np.nan, mean_stroke, inplace=True)
mydata['horsepower'].replace(np.nan, mean_horse, inplace=True)
mydata['peak-rpm'].replace(np.nan, mean_peak, inplace=True)

mydata.isnull().sum()
```

```python
mydata.dropna(subset=['price'], axis=0, inplace=True)

mydata.isnull().sum()
```

```python
mydata.head()
```

```python
mydata['city_L/100KM'] = 235/mydata['city-mpg'] #convert mpg to L/100KM #new name
#add new column to datasets

mydata['highway_L/100KM'] = 235/mydata['highway-mpg'] #convert mpg to L/100KM #new name

#two column will be added
mydata.head()
```

```python
#drop a column


mydata.drop('city-mpg', axis=1, inplace=True)

mydata.drop('highway-mpg', axis=1, inplace=True)

mydata.head()
```

**Data Normalization**

*Data Normalization

*   Simple Feature Scalling
*   Min - Max
*   Z-Score




```python
mydata['width'].head()
```

```python
mydata[['width', 'length', 'height']].head()
```

```python
mydata['width'] = mydata['width']/mydata['width'].max() #max

mydata[['width', 'length', 'height']].head()
```

```python
#to check

mydata['width'].describe()
```

```python
# min - max

mydata['length'] = (mydata['length'] - mydata['length'].min()) / (mydata['length'].max() - mydata['length'].min())

mydata[['width', 'length', 'height']].head()
```

```python
mydata['length'].describe()
```

```python
#height
#z-score

mydata['height'] = (mydata['height'] - mydata['height'].mean()) / mydata['height'].std()

mydata[['width', 'length', 'height']].head() #look at those 3 column
```

```python
mydata[['width', 'length', 'height']].describe()
```

**Data** **Binning**

```python
mydata_new1['price'].describe()
```

```python
#3 group
#normal price,mid price, high price,

price_bins = np.linspace(mydata_new1['price'].min(), mydata_new1['price'].max(), num=4)
price_bins

price_groups = ['Low Price', 'Mid Price', 'High Price']
mydata_new1['price_bins'] = pd.cut(mydata_new1['price'], price_bins, labels = price_groups, include_lowest=True)

mydata_new1[['price', 'price_bins']].head()
```

```python
norm_bins = np.linspace(mydata_new1['normalized-losses'].min(), mydata_new1['normalized-losses'].max(), num=4)
norm_bins

norm_groups = ['Low', 'Mid', 'High']
mydata_new1['normalized_bins'] = pd.cut(mydata_new1['normalized-losses'], norm_bins, labels = norm_groups, include_lowest=True)

mydata_new1[['normalized-losses', 'normalized_bins']].head()

```
