```python
import pandas as pd
#rows below
##create your own datasets, concatenate in terms of rows once, and in terms of columns once again

#create dummy datasets
dumm_data1 = {
          'Name': ['Nabila', 'Mariam', 'Anis', 'Aishah', 'Saras', 'Abid', 'Hikmah', 'Amy'],
          'Age' : [23, 32, 30, 27, 24, 25, 14, 20],
          'City': ['Kuala Lumpur', 'Kuantan', 'Shah Alam', 'Selangor', 'Temerloh', 'Gambang', 'Pekan', 'Perlis'],
          'Salary': [5000, 7000, 120000, 9000, 13000, 2000, 14000, 15000],

}

dumm_data2 = {
          'Name': ['Mimi', 'Syamina', 'Deeja', 'Balqis', 'Saad', 'Kuku', 'Amar', 'Abdul'],
          'Age' : [9, 8, 7, 2, 12, 16, 18, 19],
          'City': ['Ipoh', 'Bangi', 'SeremBan', 'Johor', 'Terengganu', 'Melaka', 'Kelantan', 'Penang'],
          'Salary': [6000, 8000, 130000, 10000, 15000, 3000, 18000, 19000]

}

#change dictionary to dataframe

df_dumm_data1 = pd.DataFrame(dumm_data1)
df_dumm_data2 = pd.DataFrame(dumm_data2)

```

```python
#concatinated rows

df_dummdata = [df_dumm_data1, df_dumm_data2]

df_dummdata12 = pd.concat(df_dummdata)

df_dummdata12.head(16)
```

```python
#reset the index/index reseting

df_dummdata12 = df_dummdata12.reset_index(drop=True)

df_dummdata12.head(16)
```

```python
#create datasets to concatinate column

dumm_data3 = {

        'Siblings':[1, 3, 4, 5, 6, 1, 3, 4, 6, 7, 3, 1, 2, 8, 9, 10],
        'Colours' :['Yellow', 'Blue', 'Red', 'Green', 'Black', 'Yellow', 'Blue', 'Red', 'Green', 'Black', 'Yellow', 'Blue', 'Red', 'Green', 'Black', 'Yellow']

}

df_dumm_data3 = pd.DataFrame(dumm_data3)
df_dumm_data3.head(16)
```

```python
#concatinate column

df_alldummdata = [df_dummdata12, df_dumm_data3]

df_alldummdata = pd.concat(df_alldummdata, axis=1) #default axis is zero

df_alldummdata.head(16)
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

col1 = np.random.normal(260, 10, 1000)
col2 = np.random.normal(160, 20, 1000)
col3 = np.random.normal(700, 5, 1000)

categorical_data = ['Ali', 'Bobo', 'Coco', 'Dede', 'Eric', 'Farah', 'Gigi', 'Huda', 'Ina', 'Jojo']

col4 = np.random.choice(categorical_data, 1000)
col5 = np.random.choice(categorical_data, 1000)

start_date = datetime(2019, 1, 1)
end_date = datetime(2020, 1, 1)

delta = end_date - start_date

date_list = [start_date + timedelta(days=x) for x in range(delta.days)]

col6 = np.random.choice(date_list, 1000)

data = np.column_stack((col1, col2, col3, col4, col5, col6))

df = pd.DataFrame(data, columns=['Subject', 'Gred', 'Class', 'Name1', 'Name2', 'Date'])

df.head()
```
