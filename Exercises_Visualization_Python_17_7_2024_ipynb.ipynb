```python
gender = ['Abu', 'Fahmi', 'Aitin', 'Mila', 'Bubu', 'Nabila']
age = [10, 12, 9, 16, 25, 35]
colors = ['skyblue', 'maroon', 'lavender', 'purple', 'silver', 'gold']

# Create a bar plot
plt.figure(figsize = (10, 6))
bars=plt.bar(gender, age, color = colors)

# Add the value of each bar on top of the bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval}', ha='center', va='bottom', fontsize=12)

plt.xlabel('Subject')
plt.ylabel('Gred')
plt.title('Harmoni Class')

plt.show()
```

```python
#create bar chart
# Data
categories = ['Dance', 'Education', 'Comedy', 'Lip Sync', 'DIY and Craft']
likes = [3500, 2800, 5000, 1500, 2000]

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, likes, color='skyblue')

# Adding titles and labels
plt.title('Likes per TikTok Video Category')
plt.xlabel('Categories')
plt.ylabel('Number of Likes')

# Add the value of each bar on top of the bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval}', ha='center', va='bottom', fontsize=12)

# Display the plot
plt.xticks(rotation=45)  # Rotate category labels for better readability
plt.tight_layout()
plt.show()
```
